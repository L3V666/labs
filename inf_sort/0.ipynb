0. Пузырек и его товарищи

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
```

```python
def linear_func(x, a, b):
    return a * x + b

def lsm(n, t): 
    params, covariance = curve_fit(linear_func, n, t)
    x = np.linspace(min(n), max(n), 1000)
    y = linear_func(x, params[0], params[1])
    return x, y, params[0], np.sqrt(np.diag(covariance))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18, 8))
f, a = plt.subplots(1, 1, figsize=(10, 6))
a.set_xlabel('n')
a.set_ylabel('t')
a.grid()
```

Пузырёк

```python
data = np.array(pd.read_csv("bubble_sort_csv.csv", header=None))

n = data[:, 0]
t = data[:, 1]

n_log = np.log(n)
t_log = np.log(t)
x, y = lsm(n_log, t_log)[0: 2]
print(lsm(n_log, t_log)[2])
k = lsm(n_log, t_log)[2]

ax[0].plot(x, y)

ax[0].scatter(n_log, t_log, marker='+', markersize=12)

ax[0].set_title("Логарифмические оси")
ax[0].set_xlabel("ln(n)")
ax[0].set_ylabel("ln(t)")
ax[0].set_title(f"bubble, k={k}")

ax[0].grid()


a.plot(n, t, label='bubble')
```
