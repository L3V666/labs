```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

fig, ax = plt.subplots(2, 1, figsize=(10, 16))
```

N^2

```python
data = np.array(pd.read_csv('bubble_sort_csv.csv', header=None))

ax[0].plot(data[:14, 0], data[:14, 1], label='c++')

data = np.array(pd.read_csv('bubble_py_csv.csv', header=None))

ax[0].plot(data[:, 0], data[:, 1], label='python')

ax[0].legend()

ax[0].set_title('N^2')
ax[0].set_xlabel('t')
ax[0].set_ylabel('n')
```

NlogN

```python
data = np.array(pd.read_csv('merge_sort_csv.csv', header=None))

ax[1].plot(data[:14, 0], data[:14, 1], label='c++')

data = np.array(pd.read_csv('merge_py_csv.csv', header=None))

ax[1].plot(data[:, 0], data[:, 1], label='python')

ax[1].legend()
ax[1].set_title('NlogN')
ax[1].set_xlabel('t')
ax[1].set_ylabel('n')
```
