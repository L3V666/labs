```python
import numpy as np
```

```python
L = 2.2
M = 2.9
g = 9.81

x = np.array([9.3, 9.8, 9.5, 9.5, 9.6]) * 1e-3
m = np.array([0.512, 0.507, 0.506, 0.509, 0.504]) * 1e-3

def u(x, m):
    return M/m * np.sqrt(g / L) * x

print(u(x, m))
print(np.mean(u(x, m)))
```

```python
x = np.array([4, 4.5, 4.2, 4.4, 4.4]) * 1e-2
m = np.array([0.502, 0.504, 0.510, 0.511, 0.504]) * 1e-3

k = 0.304
d = 141 * 1e-2
r = 20.5 * 1e-2

def u(x, m):
    return x / d * k / (m * r)

print(u(x, m))
print(np.mean(u(x, m)))
```
